```python
import plotly.io as pio
import plotly.graph_objs as go
figure= {
    "data": [
        go.Scatter(
            x=df["bh_depth"],
            y=df["aquifer_yield"],
            name="Borehole depth vs aquifer yield",
            marker=dict(
                color='indigo'
            ),
            hoverinfo="text",

        )
    ],
    "layout": go.Layout(
        barmode="stack",
        plot_bgcolor="#ffffff",
        paper_bgcolor="#ffffff",
        title={
            "text": "<b>Borehole depth vs aquifer yield</b>",
            "y": 0.93,
            "x": 0.5,
            "xanchor": "center",
            "yanchor": "top",
        },
        titlefont={"color": "black", "size": 20},
        hovermode="x",
        xaxis=dict(
            title="<b>Borehole Depth</b>",
            color="black",
            showline=True,
            showgrid=True,
            showticklabels=True,
            linecolor="black",
            linewidth=2,
            ticks="outside",
            tickfont=dict(family="Arial", size=12, color="black"),
        ),
        yaxis=dict(
            title="<b> Aquifer Yield </b>",
            color="black",
            showline=True,
            showgrid=True,
            showticklabels=True,
            linecolor="black",
            linewidth=2,
            ticks="outside",
            tickfont=dict(family="Arial", size=12, color="black"),
        ),
        legend={
            "orientation": "h",
            "bgcolor": "#ffffff",
            "xanchor": "center",
            "x": 0.5,
            "y": -0.6,
        },
        font=dict(family="sans-serif", size=12, color="black"),
    ),
}
fig = go.Figure(figure)
pio.show(fig)
```
